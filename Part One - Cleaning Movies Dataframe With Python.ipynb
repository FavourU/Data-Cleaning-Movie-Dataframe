# **Cleaning the Data**

## Remove Gross and One-line column


```python
df2.drop(columns = ['Gross', 'ONE-LINE'], axis = 'columns', inplace = True)
```

```python
df2
```

## Remove brackets from Year column

```python
df2.YEAR
```

```python
df2['YEAR'] = df2['YEAR'].str.strip('(').str.strip(')')
df2['YEAR']
```

## Index year values to keep only first year, Labeled Start Year.

```python
df2['Start Year'] = df2['YEAR'].str[:4]
df2['Start Year']
```

```python
df2
```

## Create a 'series' column based on Year column

```python
df2['char_length'] = df2['YEAR'].str.len()
df2
```

```python
df2['Series'] = np.where(df2['char_length'] > 4, True, False)
df2.head()
```

```python
#Will remove 'year' and 'char_length' columns as they are unneeded but will create a new data frame to continue with this
df3 = df2.drop(columns = ['YEAR','char_length'], axis = 'columns',)
df3
```

```python
#Will change True/ false to yes/no
df3 =df3.replace({'Series': {True : 'yes', False: 'no'}})
df3
```

## Remove /n from Genre

```python
df3.columns
```

```python
df3.GENRE
```

```python
df3['GENRE'] = df3['GENRE'].str.strip('\n')
df3
```

## Adding delimeters so that Stars can be split into into Director and Actors



```python
df3
```

```python
df3['STARS'] = df3['STARS'].str.replace('Director:\n','0 ', regex=False).str.replace('Stars:\n','1 ', regex=False)
df3
```

```python
df3['STARS'] = df3['STARS'].str.replace('\n','N/A', regex=False)
df3
```

```python
df3['STARS'] = df3['STARS'].str.replace('N/A|',' ', regex=False)
df3.head(2)
```
